```python
import sklearn.datasets
import numpy as np
import matplotlib.pyplot as plt

digits = sklearn.datasets.load_digits()
```

```python
X_all = digits.data
Y_all = digits.target

split = 0.3

length_train = int((1-split) * Y_all.size)
X_train = X_all[0:length_train]
Y_train = Y_all[0:length_train]
X_test = X_all[length_train:]
Y_test = Y_all[length_train:]
```

```python
assert Y_all.size == Y_train.size + Y_test.size, \
"Splitting didn't work properly: {} != {}".format(Y_all.size, Y_train.size + Y_test.size)
```

```python
plt.gray()
plt.matshow(digits.images[177]) #shows the 177 images
plt.show()
```

```python
def kernel(a, b, param):
    sqdist = a**2 + b**2 - 2*np.dot(a, b.T)
    return np.exp(-.5 * (1/param) * sqdist)
```

```python
print(kernel(X_test,X_test,param))
```

```python
param = 0.1
n=Y_test.size
K_ss = kernel(X_test, X_test, param)

L = np.linalg.cholesky(K_ss + 1e-15*np.eye(n))
plt.subplot(2,1,1)
plt.axis([0,11,-3,3])
plt.title("Mean over all kernels")
plt.plot(X_test,np.diagonal(L))
plt.subplot(2,1,2)
plt.axis([0,11,-3,3])
plt.title("All Kernels")
plt.plot(X_test,L)
plt.show()
```

```python
f_prior = np.dot(L, np.random.normal(size=(n,1)))
#plt.axis([0,11,-3,3])
plt.plot(X_test, f_prior)
plt.show()
```

```python
# Apply the kernel function to our training points
K = kernel(X_train, X_train, param)
L = np.linalg.cholesky(K + 0.00005*np.eye(Y_train.size))

# Compute the mean at our test points.
K_s = kernel(X_train, X_test, param)
Lk = np.linalg.solve(L, K_s)
mu = np.dot(Lk.T, np.linalg.solve(L, Y_train)).reshape((n,))

# Compute the standard deviation so we can plot it
s2 = np.diag(K_ss) - np.sum(Lk**2, axis=0)
stdv = np.sqrt(s2)
# Draw samples from the posterior at our test points.
L = np.linalg.cholesky(K_ss + 1e-6*np.eye(n) - np.dot(Lk.T, Lk))
f_post = mu.reshape(-1,1) + np.dot(L, np.random.normal(size=(n,3)))
```
